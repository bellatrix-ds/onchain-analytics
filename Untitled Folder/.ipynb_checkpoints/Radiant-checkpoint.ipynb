# Borrow Volume (Daily)

```python
df_b = df[df['event_signature']=='0xc6a898309e823ee50bac64e45ca8adba6690e99e7841c45d754e2a38e9019d9b']
df_b = df_b.rename(columns={
    'event_signature': 'Borrow_event'})
```

```python
df_b = df_b.drop(['price_oracle_raw','token','APR'],axis=1)
```

```python

decimals_map = {
    "USDC": 6,
    "USDT": 6,
    "PYUSD": 6,
    "wUSDR": 6,
    "WBTC": 8,
    "WETH": 18,
    "wstETH": 18,
    "rsETH": 18,
    "rETH": 18,
    "LINK": 18,
    "UNI": 18,
    "CBETH": 18,
    "UNKNOWN": None
}

def convert_raw_to_amount(row):
    symbol = row['symbol']
    raw_value = int(row['loan_amount'], 16)  # تبدیل hex به عدد
    decimals = decimals_map.get(symbol)
    if decimals is not None:
        return raw_value / (10 ** decimals)
    else:
        return None 

df_b['loan_amount_real'] = df_b.apply(convert_raw_to_amount, axis=1)
```

```python
df_b['APR'] = df_b['borrowRate'].apply(lambda x: int(x, 16) / 1e27)  # assuming RAY (1e27)
df_b = df_b.drop(['borrowRate'],axis=1)
```

```python
df_b = df_b.drop(['get_loan','data','token_again','variable_rate','loan_amount'],axis=1)
```

```python
df_b.head()
```

```python
df_b['symbol'].value_counts()
```

```python
df_b1 = df_b.groupby(['block_timestamp','symbol']).agg({'transaction_hash':'count','APR':'mean',
                                                                      'loan_amount_real':'sum'}).reset_index()


df_b1 = df_b1.rename(columns={'transaction_hash': 'borrow_volume','loan_amount_real':'loan_amount'})

df_b1
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df_b1['block_timestamp'] = pd.to_datetime(df_b1['block_timestamp'])

symbols = df_b1['symbol'].unique()

metrics = ['borrow_volume', 'APR', 'loan_amount']
titles = ['Borrow Volume Over Time', 'APR Over Time', 'Loan Amount Over Time']

for metric, title in zip(metrics, titles):
    plt.figure(figsize=(14, 6))

    for sym in symbols:
        df_sym = df_b1[df_b1['symbol'] == sym]
        df_sym = df_sym.sort_values('block_timestamp')
        plt.plot(df_sym['block_timestamp'], df_sym[metric], label=sym)

    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel(metric)
    plt.legend(title='Token')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

# withdraw (Daily)

```python
df_w = df[df['event_signature']=='0x3115d1449a7b732c986cba18244e897a450f61e1bb8d589cd2e69e6c8924f9f7']
df_w = df_w[df_w['symbol']=='WETH']

df_w = df_w.rename(columns={'event_signature': 'Withdraw_event'})
```

```python
df_w.head(2)
```

```python
def decode_weth_amount(data_hex):
    try:
        raw = int(data_hex, 16)
        return raw / 1e18
    except:
        return None   

df_w['withdraw_amount'] = df_w['data'].apply(decode_weth_amount)
```

```python
df_w1 = df_w.groupby(['block_timestamp','symbol']).agg({'transaction_hash':'count',
                                                                      'withdraw_amount':'sum'}).reset_index()


df_w1 = df_w1.rename(columns={'transaction_hash': 'withdraw_volume'})

df_w1
```
