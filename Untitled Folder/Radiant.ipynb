# Borrow 

```python
# Borrow events
df_b = df[df['event_signature']=='0xc6a898309e823ee50bac64e45ca8adba6690e99e7841c45d754e2a38e9019d9b']

# WETH only
df_b = df_b[df_b['token']== '0x000000000000000000000000c02aaa39b223fe8d0a0e5c4f27ead9083c756cc2']
```

```python
# remove 0x
df_b['data_clean'] = df_b['data'].str[2:]

# split
df_b['user'] = df_b['data_clean'].str[0:64].apply(lambda x: '0x' + x[-40:])
df_b['loan_amount_raw'] = df_b['data_clean'].str[64:128]
df_b['variable_rate_raw'] = df_b['data_clean'].str[128:192]
df_b['borrowRate_raw'] = df_b['data_clean'].str[192:256]

# manipulate
df_b['loan_amount'] = df_b['loan_amount_raw'].apply(lambda x: int(x, 16) / 1e18)
df_b['variable_rate'] = df_b['variable_rate_raw'].apply(lambda x: int(x, 16))
df_b['borrowRate'] = df_b['borrowRate_raw'].apply(lambda x: int(x, 16) / 1e27)
df_b['APR'] = df_b['borrowRate'] * 100
```

```python
df_b.head(2)
```

```python
df_b = df_b.drop(['event_signature','token','get_loan','data','data_clean','variable_rate_raw','borrowRate_raw','price_oracle_raw',
                 'loan_amount_raw'],axis=1)

#df_b1 = df_b1.rename(columns={'transaction_hash': 'borrow_volume','loan_amount_real':'loan_amount'})
```

```python
borrow = df_b.groupby(['block_timestamp']).agg({'transaction_hash':'count','loan_amount':'sum',
                                               'borrowRate':'mean', 'APR':'mean' }).reset_index()

borrow = borrow.rename(columns={'transaction_hash': 'borrow_volume'})

#borrow
```

# Deposit

```python
# Deposit events
df_d = df[df['event_signature']=='0xde6857219544bb5b7746f48ed30be6386fefc61b2f864cacf559893bf50fd951']

# WETH only
df_d = df_d[df_d['token']== '0x000000000000000000000000c02aaa39b223fe8d0a0e5c4f27ead9083c756cc2']
```

```python
# remove 0x
df_d['data_clean'] = df_d['data'].str[2:]  

df_d['Aggregator'] = df_d['data_clean'].str[:64].apply(lambda x: '0x' + x[-40:])     
df_d['deposit_raw_amount'] = df_d['data_clean'].str[64:]

df_d['deposit_amount'] = df_d['deposit_raw_amount'].apply(lambda x: int(x, 16) / 1e18)
```

# Withdraw

```python
# Withdraw events
df_w = df[df['event_signature']=='0x3115d1449a7b732c986cba18244e897a450f61e1bb8d589cd2e69e6c8924f9f7']

# WETH only
df_w = df_w[df_w['token']== '0x000000000000000000000000c02aaa39b223fe8d0a0e5c4f27ead9083c756cc2']
```

```python
def extract_withdraw_amount(data_hex):
    if pd.isna(data_hex):
        return None
    hex_clean = data_hex[2:]   
    amount_hex = hex_clean[-64:]   
    amount = int(amount_hex, 16) / 1e18 
    return amount

df_w['withdraw_amount'] = df_w['data'].apply(extract_withdraw_amount)
```

```python
withdraw = df_w.groupby(['block_timestamp']).agg({'transaction_hash':'count','withdraw_amount':'sum'}).reset_index()

withdraw = withdraw.rename(columns={'transaction_hash': 'withdraw_volume'})

#withdraw
```

# Repay

```python
# Repay events
df_r = df[df['event_signature']=='0x4cdde6e09bb755c9a5589ebaec640bbfedff1362d4b255ebf8339782b9942faa']

# WETH only
df_r = df_r[df_r['token']== '0x000000000000000000000000c02aaa39b223fe8d0a0e5c4f27ead9083c756cc2']
```

```python
def decode_repay_amount(data_hex):
    if pd.isnull(data_hex) or not isinstance(data_hex, str) or len(data_hex) < 66:
        return None
    data_clean = data_hex[2:] if data_hex.startswith('0x') else data_hex
    repay_amount_hex = data_clean[-64:]  
    repay_amount = int(repay_amount_hex, 16) / 1e18      
    return repay_amount

df_r['repay_amount'] = df_r['data'].apply(decode_repay_amount)
```

```python
df_r.head(2)
```

```python
repay = df_r.groupby(['block_timestamp']).agg({'transaction_hash':'count','repay_amount':'sum'}).reset_index()

repay = repay.rename(columns={'transaction_hash': 'repay_volume'})

repay
```

# Final df

```python
from functools import reduce
import pandas as pd


for df in [borrow, deposit ,withdraw , repay]:
    df['block_timestamp'] = pd.to_datetime(df['block_timestamp'])

dfs = [borrow, deposit ,withdraw , repay]

final_df = reduce(lambda left, right: pd.merge(left, right, on=['block_timestamp'], how='outer'), dfs)

final_df = final_df.sort_values('block_timestamp').reset_index(drop=True)
```

```python
new_order = [ 'block_timestamp', 'deposit_amount','loan_amount', 'repay_amount', 'withdraw_amount',
             'APR','deposite_volume' ,'borrow_volume','repay_volume','withdraw_volume']

final_df = final_df[new_order]

final_df = final_df.fillna(0)
```

```python
final_df
```

# 1

```python
final_df['utilization_rate'] = final_df['loan_amount'] / (final_df['deposit_amount'] + final_df['loan_amount'])
final_df['net_flow'] = final_df['deposit_amount'] - final_df['withdraw_amount']
final_df['liquidity_delta'] = final_df['deposit_amount'] + final_df['repay_amount'] - final_df['loan_amount'] - final_df['withdraw_amount']
final_df['debt_outstanding'] = (final_df['loan_amount'] - final_df['repay_amount']).cumsum()
final_df['apr_change'] = final_df['APR'].diff()
```
